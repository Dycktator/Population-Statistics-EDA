```python
import pandas as pd
import seaborn as sns
sns.set_theme(style="ticks", font_scale=1.3)
import matplotlib.pyplot as plt
plt.rc("axes.spines", top=False, right=False)
plt.figure(figsize = (20,20))
dfu = pd.read_csv("Vancouver Population Statistics.csv")
dfu = dfu.drop(columns=['Note', "Flag_Total", "Flag_Male", "Male", "Female"])
dfu = dfu.drop([0])
dfu.describe().T
```

# 1) Commuting Distances

#### Commuting destination for the employed labour force aged 15 years and over in private households with a usual place of work.

```python
df0 = dfu.loc[2:5,]
df0 = df0.apply(lambda x: pd.to_numeric(x) if x.name=='Total' else x)
df0.loc[2,"Topic"]="CD1"
df0.loc[3,"Topic"]="CD2"
df0.loc[4,"Topic"]="CD3"
df0.loc[5,"Topic"]="CD4"
df0
```

```python
sns.catplot(x="Topic" , y="Total", kind="bar", hue="Characteristics", data=df0)
plt.title('Commuting Distances')
plt.xlabel('')
plt.ylabel('Total %')
plt.show()
```
